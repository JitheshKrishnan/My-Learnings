```python
x = np.linspace(0, 10, 50)
x
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
plt.figure(figsize = (5,5))
plt.scatter(x, y) #scatter plot
plt.show()
```

```python
plt.scatter(y, x)
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x, y, c = "y", marker = "*", s = 300) # or plt.scatter(x, y, color="r") or "color ='#987710'
plt.show() # marker plots in that symbol, s = 100 indicates the size of the plotting symbol
```

```python
plt.plot(x, z, c = 'g', linestyle = '--', marker = '^', markersize = '8')
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x, y, c = "r", marker = "*", s = 500, label = 'Sin')
plt.plot(x, z, c = "g", marker = "^", markersize = 20, linestyle = '--', label = 'Cos')# or linestyle='dotted'
plt.title("Sin and Cos", fontsize = 20)
plt.xlabel("X-axis", fontsize = 20)
plt.ylabel("Y-axis", fontsize = 20)
plt.xticks(range(10)) #x-axis range labelling
#plt.yticks([-1, 1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75 ]) #y-axis range labelling
plt.xlim([0, 2*np.pi]) #x-axis limit setting
plt.ylim([-1, 1]) #y-axis limit setting
plt.legend() #only if you put this fn the label of each curve will be displayed
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize = (16,9))

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
#if a var like ax1 is used then we should add set- in xlabels

ax1.plot(x, z, color = "g", marker = "<", markersize = 20, linestyle = '--', label = 'Cos')# or linestyle='dotted'
ax1.set_title("Sin", fontsize = 20)
ax1.set_xlabel("X-axis", fontsize = 10)
ax1.set_ylabel("Y-axis", fontsize = 10)

ax2.scatter(x, y, c = "r", marker = "*", s = 500, label = 'Sin')
ax2.set_title("Cos", fontsize = 20)
ax2.set_xlabel("X-axis", fontsize = 10)
ax2.set_ylabel("Y-axis", fontsize = 10)

ax3.plot(x, z, color = "r", marker = "o", markersize = 10, linestyle = '--', label = 'Cos')# or linestyle='dotted'
ax3.set_title("Sin", fontsize = 20)
ax3.set_xlabel("X-axis", fontsize = 10)
ax3.set_ylabel("Y-axis", fontsize = 10)

ax4.scatter(x, y, color = "g", marker = "s", s = 10, label = 'Cos')# or linestyle='dotted'
ax4.set_title("Sin", fontsize = 20)
ax4.set_xlabel("X-axis", fontsize = 10)
ax4.set_ylabel("Y-axis", fontsize = 10)

plt.show()


```

# BAR GRAPHS

```python
marks = [90, 34, 56, 72, 64, 30]
names = ['naren', 'jithesh', 'aslam', 'priya', 'nandy', 'vishnu']

plt.bar(names, marks, color = 'g', width = 0.5)
plt.title("Marks class 10")
plt.xlabel("Names-->")
plt.ylabel("Marks-->")

plt.show()
```

```python
x = np.random.normal(0, 4, 10000) # normal distribution are concentrated more towards the mean
z = np.random.uniform(0, 4, 10000) # in uniform, 0 is the smallest value and 4 is the largest value

plt.hist(x, bins = 200, density = True, alpha = 0.6, color = 'r')#bins is the bars
plt.grid(True)
plt.show()
```

```python
plt.hist(z, bins = 50, color = 'b') #bins are the bars
plt.show()
```

```python
ice_creams = [30, 23, 12, 21, 4, 10]
names = ["mango", "chocobar", "pine apple", "orange", "butterscotch", "rabdi kulfi"]

plt.pie(ice_creams, labels = names)
plt.show()
```
