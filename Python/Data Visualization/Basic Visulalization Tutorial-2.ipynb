```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(0,10,50)
x
```

```python
y=np.sin(x)
y
```

```python
z=np.cos(x)
z
```

```python
w=np.tan(x)
w
```

```python
plt.figure(figsize=(16,9))
plt.plot(x,y,marker='^',c='g',linestyle='--',markersize=10,label='sin')
plt.scatter(x,z,marker='s',c='y',s=30,label='cos')
plt.xlabel('x values-->',fontsize=20)
plt.ylabel('anuja',fontsize=20)
plt.title('sin wave',fontsize=30)
plt.xlim(0,8)
plt.ylim(-5,5)
plt.xticks(range(0,10))
plt.yticks(range(-2,2))
plt.grid(c='r',linestyle='dotted')
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(16,18))

a1=fig.add_subplot(2,2,1)
a2=fig.add_subplot(2,2,2)
a3=fig.add_subplot(2,2,3)
a4=fig.add_subplot(2,2,4)

a1.plot(x,y,c='g',marker='^',markersize=20,linestyle='--',label='sin')
a1.set_xlabel('x values-->',fontsize=20)
a1.set_ylabel('y values-->',fontsize=20)
a1.set_title('sin wave',fontsize=30)

a2.plot(x,z,c='r',marker='^',markersize=20,linestyle='--',label='cos')
a2.set_xlabel('x values-->',fontsize=20)
a2.set_ylabel('y values-->',fontsize=20)
a2.set_title('cos wave',fontsize=30)

a3.plot(x,y,c='y',marker='*',markersize=20,linestyle='--',label='sin')
a3.set_xlabel('x values-->',fontsize=20)
a3.set_ylabel('y values-->',fontsize=20)
a3.set_title('sin wave',fontsize=30)

a4.plot(x,z,c='black',marker='o',markersize=20,linestyle='--',label='cos')
a4.set_xlabel('x values-->',fontsize=20)
a4.set_ylabel('y values-->',fontsize=20)
a4.set_title('cos wave',fontsize=30)

plt.show()
```
