```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline 
#include this statement so no need to use plot function each time
auto = pd.read_csv('Automobile[1].csv')
auto.head()
```

```python
sns.distplot(auto['horsepower'])
plt.show()
```

```python
sns.distplot(auto['weight'])
plt.show()
```

```python
sns.displot(auto['horsepower'], kde = False, rug = True)
plt.show()
```

```python
sns.histplot(auto['horsepower'], kde = True, color = 'pink')
plt.show()
```

```python
sns.jointplot(data = auto, x = auto['horsepower'], y = auto['weight'])
plt.show()
```

```python
sns.jointplot(x = auto['horsepower'], y = auto['weight'], data = auto, color = 'r', kind = "hex")
plt.show()
```

```python
sns.jointplot(x = auto['horsepower'], y = auto['weight'], data = auto, color = 'g', kind = "resid")
plt.show()
```

```python
sns.jointplot(x = auto['horsepower'], y = auto['weight'], data = auto, color = 'black', kind = "reg")
plt.show()
```

```python
sns.jointplot(x = auto['horsepower'], y = auto['weight'], data = auto, color = 'orange', kind = "kde")
plt.show()
```

```python
sns.pairplot(auto[['horsepower', 'weight', 'displacement']], kind = 'kde')
plt.show()
```

```python
sns.pairplot(auto[['horsepower', 'weight', 'displacement']])
```

```python
sns.pairplot(auto[['horsepower', 'weight']])
```

```python
plt.figure(figsize = (9,9))
sns.stripplot(data = auto, x = auto['origin'], y = auto['horsepower'], color = 'red', jitter = True)
plt.show()
```

```python
plt.figure(figsize = (9,9))
sns.stripplot(data = auto, x = auto['origin'], y = auto['horsepower'], color = 'red', jitter = False)
plt.show()
```

```python
plt.figure(figsize = (9,9))
sns.swarmplot(data = auto, x = auto['origin'], y = auto['horsepower'], color = 'brown')
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.boxplot(data = auto, x = auto['origin'], y = auto['horsepower'], hue = auto['model_year'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.boxplot(data = auto, x = auto['origin'], y = auto['horsepower'], hue = auto['cylinders'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.violinplot(data = auto, x = auto['origin'], y = auto['horsepower'], hue = auto['cylinders'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.barplot(data = auto, x = auto['origin'], y = auto['horsepower'], hue = auto['model_year'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.barplot(data = auto, x = auto['origin'], y = auto['horsepower'], hue = auto['cylinders'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.countplot(x = auto['origin'], hue = auto['cylinders'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.pointplot(x = auto['cylinders'], y = auto['horsepower'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.pointplot(x = auto['cylinders'], y = auto['horsepower'], hue = auto['origin'])
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.catplot(x = 'cylinders', y = 'horsepower', hue = 'acceleration', col = 'origin', data = auto, kind = 'point')
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.catplot(x = 'cylinders', y = 'horsepower', hue = 'acceleration', col = 'origin', data = auto, kind = 'swarm')
plt.show()
```
