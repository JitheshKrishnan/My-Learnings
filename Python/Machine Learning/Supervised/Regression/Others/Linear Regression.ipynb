```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,9)
y = np.array([1.5, 3.2, 7, 7.5, 9, 11.9, 15, 15.8])
```

```python
plt.plot(x, y, marker = '*')
plt.show()
```

```python
w1, w0 = np.random.rand(2)
w1, w0
```

```python
# using yhat = w1*x + w0 

yhat = []
for xi in x:
    yhat.append(w1*xi+w0)
    
plt.plot(x, y, marker = '*', label  = 'Original Data')
plt.plot(x, yhat, c = 'r', label = 'Initial Random Line')
plt.legend()
plt.show()
```

```python
lr = 0.001
epoch = 1000
error = []
```

```python
for i in range(epoch):
    epoch_error = 0
    # calculating the error
    for i in range(len(x)):
        y_pred = w1*x[i]+w0
        epoch_error+= (y[i]-y_pred)**2
    error.append(epoch_error)
    
    dw1, dw0 = 0, 0
    # calculating the partial derivative
    for i in range(len(x)):
        dw1+= -2*x[i]*(y[i]-(w1*x[i]+w0)) # -2*x*(y-yhat)
        dw0+= -2*(y[i]-(w1*x[i]+w0)) # -2*(y-yhat)
    # update the weight:
    w1 = w1 - lr*dw1
    w0 = w0 - lr*dw0
    
    print(f'w1 = {w1}, w0 = {w0}')
```

```python
error
```

```python
# using yhat = w1*x + w0 

yp = []
for xi in x:
    yp.append(w1*xi+w0)
    
plt.plot(x, y, marker = '*', label  = 'Original Data')
plt.plot(x, yhat, c = 'r', label = 'Initial Random Line')
plt.plot(x, yp, c = 'g', label = 'Fitted line')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16, 9))
plt.bar(range(1,1001), error)
plt.ylim(0, 20)
plt.xlim(0, 10)
plt.show()
```
