```python
X = df["FUEL CONSUMPTION COMB (L/100 km)"]
y = df["CO2EMISSIONS"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
10, 1
```

```python
X_train = np.array(X_train).reshape(-1, 1) # do not do this step if you are using more column than 1
X_test = np.array(X_test).reshape(-1, 1)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
model = reg.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
plt.scatter(X_test, y_test, c = "g", label = "Actual")
plt.scatter(X_test, y_pred, c = "r", label = "predicted")
plt.legend()
plt.show()
```
