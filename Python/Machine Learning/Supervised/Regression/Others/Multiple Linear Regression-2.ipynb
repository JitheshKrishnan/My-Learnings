```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Automobile+%281%29.csv')
df.dropna(axis = 0, inplace = True)
df1 = df[['symboling', 'normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_size', 'bore', 'stroke',
  'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']]

x = df.drop('price', axis = 1)
y = df['price']
cat_col = ['make', 'fuel_type', 'aspiration', 'number_of_doors', 'body_style', 'drive_wheels', 'engine_location', 'engine_type', 'number_of_cylinders', 'fuel_system']

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cat_col:
    le.fit(x[col])
    x[col] = le.transform(x[col])

from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit(x)
mm.transform(x)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.3, random_state = 39)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X_train, y_train)
y_pred = model.predict(X_test)

plt.scatter(x = X_test.iloc[:, 0], y = y_test, color = 'g', label = 'Actual')
plt.scatter(x = X_test.iloc[:, 0], y = y_pred, color = 'r', label = 'Predicted')
plt.xlabel('X_test -->')
plt.ylabel('y_pred/y_test -->')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
