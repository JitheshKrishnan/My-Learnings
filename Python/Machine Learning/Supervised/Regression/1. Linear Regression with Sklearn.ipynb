```python
x = df[['normalized_losses']]
y = df['price']
```

```python
x, y
```

```python
# not doing label encoding as it is a simple linear regression using a single feature
# but we are normalizing it
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
```

```python
mm.fit(x)
```

```python
x = mm.transform(x)
```

```python
x
```

```python
x[:,0].max()
```

```python
x[:,0].min()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
plt.scatter(X_test, y_test, c = "g", label = "Actual")
plt.scatter(X_test, y_pred, c = "r", label = "predicted")
plt.title('Actual vs Predicted')
plt.xlabel('Input values -->')
plt.ylabel('Output values -->')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
