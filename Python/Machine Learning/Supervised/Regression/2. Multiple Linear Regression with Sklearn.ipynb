```python
x = df.drop('price', axis = 1)
y = df['price']
```

```python
x
```

```python
cat_col = ['make', 'fuel_type', 'aspiration', 'number_of_doors', 'body_style', 'drive_wheels', 'engine_location', 'engine_type', 'number_of_cylinders', 'fuel_system']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in cat_col:
    le.fit(x[col])
    x[col] = le.transform(x[col])
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
```

```python
mm.fit(x)
x = mm.transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.3, random_state = 39)
```

```python
a = [X_train, X_test, y_train, y_test]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
sns.regplot(x = y_test, y = y_pred, data = a, color = 'r')
plt.xlabel('y_test -->')
plt.ylabel('y_pred -->')
plt.title('Actual vs Predicted')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
