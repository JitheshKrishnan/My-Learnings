# Pandas Visualization

```python
dict = { 'name':['n1','n2','n3','n4','n5','n6'],
              'age':[20,20,21,20,21,20],
              'math_marks':[100,90,91,98,92,95],
              'physics_marks':[90,100,91,92,98,95],
              'chem_marks' :[93,89,99,92,94,92]
              }
df = pd.DataFrame(dict)
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind = 'scatter', x = 'name', y = 'age', color = 'blue')
plt.title('Scatter Plot')
plt.show()
```

```python
df.plot(kind = 'bar', x = 'name', y = 'math_marks', color = 'red')
plt.title('Bar Plot')
plt.show()
```

```python
# Get Current Axis
ax = plt.gca()

df.plot(kind = 'line', x = 'name', y = 'math_marks', color = 'red', ax = ax)
df.plot(kind = 'line', x = 'name', y = 'physics_marks', color = 'blue', ax = ax)
df.plot(kind = 'line', x = 'name', y = 'chem_marks', color = 'green', ax = ax)
plt.title('Line Plot')
plt.grid()
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(20, 4), columns = ['A', 'B', 'C', 'D'])

df.plot.area()
plt.show()
```

```python
df.sum().plot.barh(color = 'purple')
plt.title('Horizontal Bar Plot')
plt.show()
```

```python
df = pd.DataFrame(
        { 
            "A": np.sqrt(np.random.randn(1000) +1),
            "B": np.random.randn(1000),
        },
        columns = ["A","B"],
    )
df.plot.hist(alpha = 0.5)
plt.title('Histogram Plot')
plt.xlabel('Columns')
plt.grid()
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(30, 4),columns=["A","B","C","D"])

df.plot.box()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(100, 2), columns=["a", "b"])
df["b"] = df["b"] + np.arange(100)
df.plot.hexbin(x = 'a', y = 'b', gridsize = 15, color = 'yellow', cmap = 'YlGnBu')
plt.show()
```

```python
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="Pie plot")
series.plot.pie(figsize = (6, 6))
plt.show()
```
