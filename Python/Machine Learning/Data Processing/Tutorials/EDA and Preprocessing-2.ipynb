```python
nvc = ['normalized-losses', 'num-of-doors', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
for i in nvc:
    df[i] = df[i].replace('?', np.NaN)
    df[i] = pd.to_numeric(df[i], errors = 'coerce')
    mean = df[i].mean(skipna = True)
    df[i] = df[i].replace(np.NaN, mean)
df.isna().sum()
```

```python
plt.figure(figsize = (16,9))
sns.countplot(data = df, x = 'highway-mpg')
plt.show()
```

```python
sns.distplot(df['highway-mpg'], kde = True, color = 'green')
plt.show()
```
