```python
import psutil
import time
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
fig, axes = plt.subplots(2, 1)
fig.show()
```

```python
i = 0
x, y = [], []
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    axes[0].plot(x, y, color = 'g')
    axes[0].set_xlim(left = max(0, i-50), right = i-50)
    sns.lineplot(x = x, y = y, ax = axes[1], color = 'r')
    axes[1].set_xlim(left = max(0, i-50), right = i-50)
    fig.canvas.draw()
    time.sleep(0.1)
    i+=1
```
