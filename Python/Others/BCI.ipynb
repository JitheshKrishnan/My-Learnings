```python
#!pip install openpyxl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = pd.read_excel('BrainFlow-RAW_blink-1-jk_6.xlsx')
```

```python
x = pd.DataFrame(x)
x
```

```python
cols = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x']
x.columns = cols
```

```python
x
```

```python
c = x.index
```

```python
b = x['q']
b
```

```python
plt.figure(figsize = (16,9))
plt.plot(c, b, c = 'b', label = 'EEG')
plt.title('Blink signals', fontsize = 25)
plt.xlabel('Time -->', fontsize = 20)
plt.ylabel('Micro Volts -->', fontsize = 20)
plt.xlim([0, 300])
plt.ylim([-10, 300])
plt.grid(c = 'r', linestyle = '--')
plt.legend()
plt.show()
```

```python
x.plot()
plt.show()
```

```python
a = np.linspace(0, 50, 5000)
x = np.sin(a)
y = np.cos(a)
z = np.tan(a)

fig = plt.figure(figsize = (16,9))
plot1 = fig.add_subplot(2, 2, 1)
plot2 = fig.add_subplot(2, 2, 2)
plot3 = fig.add_subplot(2, 2, 3)
plot4 = fig.add_subplot(2, 2, 4)

plot1.plot(a, x, c = 'red', linestyle = 'dotted', label = 'sin')
plot1.set_title('Sine wave', fontsize = '10')
plot1.set_xlabel('input value a -->', fontsize = '5')
plot1.set_ylabel('output value x -->', fontsize = '5')
plot1.grid(c = 'blue')
plot1.legend()

plot2.plot(a, y, c = 'green', linestyle = '--', label = 'cos')
plot2.set_title('Cosine wave', fontsize = '10')
plot2.set_xlabel('input value a -->', fontsize = '5')
plot2.set_ylabel('output value y -->', fontsize = '5')
plot2.grid(c = 'red')
plot2.legend()

plot3.scatter(a, z, c = 'blue', marker = '*', s = 10, label = 'tan')
plot3.set_title('Tan wave - 1', fontsize = '10')
plot3.set_xlabel('input value a -->', fontsize = '5')
plot3.set_ylabel('output value z -->', fontsize = '5')
plot3.set_xlim([0, 30])
plot3.set_ylim([-100, 100])
plot3.grid(c = 'orange')
plot3.legend()

plot4.scatter(a, z, c = 'purple', marker = 's', s = 10, label = 'tan')
plot4.set_title('Tan wave - 2', fontsize = '10')
plot4.set_xlabel('input value a -->', fontsize = '5')
plot4.set_ylabel('output value z -->', fontsize = '5')
plot4.set_xlim([0, 15])
plot4.set_ylim([-10, 10])
plot4.grid(c = 'pink')
plot4.legend()

plt.show()
```
