```python
brain = cv2.imread('brain.jpg')
brain = cv2.cvtColor(brain, cv2.COLOR_BGR2RGB)
plt.imshow(brain)
plt.show()
```

```python
b, g, r = cv2.split(brain)

fig = plt.figure(figsize = (16, 9))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.imshow(r, cmap = 'gray')
ax1.set_title('RED')

ax2.imshow(g, cmap = 'gray')
ax2.set_title('GREEN')

ax3.imshow(b, cmap = 'gray')
ax3.set_title('BLUE')

plt.show()
```

```python
hsv = cv2.cvtColor(brain, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

fig = plt.figure(figsize = (16, 9))

ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.imshow(h, cmap = 'viridis')
ax1.set_title('HUE')

ax2.imshow(s, cmap = 'gray')
ax2.set_title('SATURATION')

ax3.imshow(v, cmap = 'gray')
ax3.set_title('lightness')

plt.show()
```

```python
gray_brain = cv2.cvtColor(brain, cv2.COLOR_BGR2GRAY)
#cv2_imshow(gray image)

gray_brain = cv2.cvtColor(gray_brain, cv2.COLOR_BGR2RGB)
plt.title("Grayscale Image") 
plt.imshow(gray_brain)
plt.show()
```

```python
resl = cv2.resize(image, (500, 420), interpolation = cv2.INTER_AREA)
resh = cv2.resize(image, (3000, 2550), interpolation = cv2.INTER_AREA)

fig = plt.figure(figsize = (16, 9))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.imshow(image)
ax1.set_title('Input Image')

ax2.imshow(resl)
ax2.set_title('Scaled Down Image')

ax3.imshow(resh)
ax3.set_title('Scaled Up Image')

plt.show()
```

```python
tumor = image[200:500, 450:750]
plt.imshow(tumor)
plt.title('Tumor Part')
plt.show()
```

```python
vflip = cv2.flip(image, 0)
hflip = cv2.flip(image, 1)

fig = plt.figure(figsize = (16, 9))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)

ax1.imshow(vflip)
ax1.set_title('Vertically Flipped Image')

ax2.imshow(hflip)
ax2.set_title('Horizontally Flipped Image')

plt.show()
```

```python
trp = cv2.transpose(image)
print('Original Shape: ', image.shape)
print('Transpose Image Shape: ', trp.shape)
plt.imshow(trp)
plt.title('Transposed Image')
plt.show()
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
h += 20
s += 16
finalhsv = cv2.merge((h, s, v))
img = cv2.cvtColor(finalhsv, cv2.COLOR_HSV2BGR)
newimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize = (16, 9))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.imshow(image)
ax1.set_title('Input Image')

ax2.imshow(newimg)
ax2.set_title('Processed Image')

plt.show()
```
